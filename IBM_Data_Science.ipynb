### ***Session 4 : Data Cleaning***

*<font color='lightgreen'>i) Filter Out Useful Features*

```python
data.drop(columns=['id','yr_renovated','lat','long','sqft_living15','sqft_lot15'],axis=1,inplace=True)
```

*<font color='lightgreen'>ii) Data Formatting*

```python
data['date']=pd.to_datetime(data['date']).dt.date
data['date']=pd.to_datetime(data['date'])
```

```python
data['bathrooms']=round((data['bathrooms'])).astype('Int64')
data['floors']=round((data['floors'])).astype('Int64')
```

```python
data['year_sold']=data['date'].dt.year
data['month_sold']=data['date'].dt.month
data['day_sold']=data['date'].dt.day
data['age']=np.subtract(data['year_sold'],data['yr_built']).astype('Int64')
```

```python
data.replace({
    'age':-1
},np.NAN,inplace=True)

data.dropna(axis=0,inplace=True)
```

```python
data.reset_index(inplace=True)
data.drop(['index'],axis=1,inplace=True)
```

*<font color='lightgreen'>iii) Rename Columns*

```python
data.rename(columns={
    'sqft_living':'living_area(sq.feet)',
    'sqft_above':'above_area(sq.feet)',
    'sqft_lot':'lot_area(sq.feet)',
    'floors':'house_floor',
    'sqft_basement':'bsmt_area(sq.feet)',
    'yr_built':'year_built'
    },inplace=True)
```

*<font color='lightgreen'> iv) Final Output/DataFrame*

```python
data=data[['house_floor','bedrooms','bathrooms','living_area(sq.feet)','above_area(sq.feet)','bsmt_area(sq.feet)','lot_area(sq.feet)','waterfront','view','grade','condition','zipcode','year_built','year_sold','month_sold','day_sold','price']]
```

```python
data.iloc[:5]
```

### ***Session : Model Development***

*<font color='lightgreen'>i) Import Libraries*

```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
import xgboost as xgb 
from sklearn.preprocessing import StandardScaler
```

*<font color='lightgreen'>ii) Split Data into Independent and Dependent Variables*

```python
X=data.drop(['price'],axis=1)
y=data['price']
```

```python
X_train,X_val,y_train,y_val=train_test_split(X,y,test_size=0.2)
```

```python
print(f'Training Dataset : {X_train.shape[0]:,.0f} rows\nValidation Dataset : {X_val.shape[0]:,.0f} rows')
```

```python
scaler=StandardScaler().fit(X)
X_train=scaler.fit_transform(X_train)
X_val=scaler.transform(X_val)
```

*<font color='lightgreen'>iii) Initialize & Fit the Model*

```python
#param_grid={'max_depth':[3,4,5],'gamma':[0.25,1.0,10.0],'learning_rate':[0.05,0.1,0.5]} #1st GridSearchCV
#max_depth=5,gamma=0.25,learning_rate=0.1

param_grid={
    'max_depth':[5],
    'gamma':[1.0,10.0],
    'learning_rate':[0.1,0.5],
}
#max_depth=5,gamma=1.0,learning_rate=0.1
```

```python
optional_param=GridSearchCV(
    estimator=xgb.XGBRegressor(n_estimators=500),
    param_grid=param_grid,
    cv=5,
    n_jobs=5,
    verbose=2
)

optional_param.fit(X_train,y_train,early_stopping_rounds=5,eval_set=[(X_val,y_val)])
print(f'Best Parameters : {optional_param.best_estimator_}')
```

```python
model=xgb.XGBRegressor(max_depth=5,gamma=1.0,learning_rate=0.1,n_estimators=500)
model.fit(X_train,y_train,early_stopping_rounds=5,eval_set=[(X_val,y_val)],verbose=True)
```

*<font color='lightgreen'>iv) Model Evaluation*

```python
y_pred=model.predict(X_val)
```

```python
print(f'R2_Score : {r2_score(y_val,y_pred)*100:,.2f}%\nMean Absolute Error : {mean_absolute_error(y_val,y_pred):,.2f}\nMean Squared Error : {mean_squared_error(y_val,y_pred):,.2f}')
```

```python
prediction=pd.DataFrame({
    'Actual':y_val,
    'Predicted':y_pred
})
prediction.iloc[:5]
```
